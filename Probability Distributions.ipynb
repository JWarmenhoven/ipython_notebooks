```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import gridspec
from scipy.stats import beta, norm, t, pareto, dweibull, alpha, gamma

%matplotlib inline
```

```python
def setup_fig(size):
    fig = plt.figure(figsize=size)
    gs = gridspec.GridSpec(4,2)
    ax1 = plt.subplot(gs[0,:-1])
    ax2 = plt.subplot(gs[1:,:-1])
    ax3 = plt.subplot(gs[:,1:])
    return(ax1, ax2, ax3)

def plot_distribution(rv, size=100, colormap=plt.cm.BrBG, figsize=(6,3)):
        
    ax1, ax2, ax3 = setup_fig(figsize)
    
    x = np.linspace(rv.ppf(0.01), rv.ppf(0.99), size)
    im = ax3.imshow(rv.rvs(size=(size,size)), cmap=colormap)
    plt.colorbar(mappable=im, ax=ax3, cax=ax1, orientation='horizontal')
    ax2.plot(x, rv.pdf(x), '-y', lw=2)
    
    plt.tight_layout();
```

## Distributions

### Beta

```python
a, b = .99, .99
rv  = beta(a, b)

plot_distribution(rv)
```

### Normal

```python
l, s  = 0, 1
rv = norm(l,s)

plot_distribution(rv)
```

### Student T

```python
df, l, s  = 2.74, 0, 1
rv = t(df,l,s)

plot_distribution(rv)
```

### Pareto

```python
b, l, s  = 2.62, 0, 1
rv = pareto(b,l,s)

plot_distribution(rv)
```

### Double Weibull

```python
c, l, s  = 2.07, 0, 1
rv = dweibull(c,l,s)

plot_distribution(rv)
```

### Alpha

```python
a, l, s  = 3.57, 0, 1
rv = alpha(a,l,s)

plot_distribution(rv)
```

### Gamma

```python
a, l, s  = 1.99, 0, 1
rv = gamma(a,l,s)

plot_distribution(rv)
```
