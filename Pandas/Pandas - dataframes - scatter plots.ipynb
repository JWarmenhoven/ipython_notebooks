```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pandas.tools.plotting as pplt

pd.options.display.mpl_style = 'default'

%matplotlib inline
pd.set_option('display.notebook_repr_html', False)
```

```python
x = np.random.random_integers(1, 10, size=40)
y = np.random.random_integers(1, 10, size=40)
size = np.random.random_integers(1, 50, size=40)**2
```

```python
x, y, size
```

<br>Een dictionary maken van de 3 variabelen. De dictionary is input voor de dataframe constructor.

```python
data = {'kolom1': x, 'kolom2': y, 'kolom3' : size}
```

```python
data
```

```python
df = pd.DataFrame(data)
```

### Scatter plot (Matplotlib)

```python
plt.scatter(df.kolom1, df.kolom2, df.kolom3, alpha=0.5)
plt.grid()
plt.title('Matplotlib Scatter')
```

### Scatter plot (Pandas) --> Een wrapper voor Matplotlib

```python
df.plot(kind='scatter', x='kolom1', y='kolom2', s=df.kolom3, title='Pandas Scatter')
```
